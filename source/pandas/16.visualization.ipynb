## 16. Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
### dice histogram

df_dice = pd.DataFrame(
    np.random.randint(1, 7, 5000),
    columns = ['Hist']
)
df_dice['Odds'] = df_dice['Hist'] + np.random.randint(1, 7, 5000)
ax = df_dice.plot.hist(bins = 12, alpha = 0.5)
```

```python
ser = pd.Series(np.random.randn(1000), 
                index=pd.date_range('11/15/2017', periods = 1000))
ser = ser.cumsum()
ser.plot()
```

```python
df_vis = pd.DataFrame(np.random.randn(1000, 3), 
                index=pd.date_range('11/15/2017', periods = 1000)
               #        ,
               # columns = list('ABC')
                     )
ser = ser.cumsum()
ser.plot()
```

```python
df_vis = pd.DataFrame(np.random.randn(1000, 3), 
                index=pd.date_range('11/15/2017', periods = 1000),
               columns = list('ABC'))
df_vis = df_vis.cumsum()
df_vis.plot()
```

```python
### bar chart
pd.DataFrame(np.random.randn(5))
pd.DataFrame(np.random.randn(5)).plot.bar()
```

```python
values = [ 'A', 'B', 'C', 'D', 'E' ]
df_rand = pd.DataFrame(np.random.rand(10, 4)).plot.bar()
```

```python
values = [ 'A', 'B', 'C', 'D']
df_rand = pd.DataFrame(np.random.rand(10, 4), columns = values)
df_rand.plot.bar()
```

```python
x_range = range(1, 15)
y_values = [1, 5, 4, 7, 6, 9, 5, 7, 10, 14, 10, 12, 9, 8]
```

```python
plt.fill_between(x_range, y_values, color='skyblue', alpha=0.5)
plt.fill()
```

```python
### area plot
pd.DataFrame(np.random.rand(10, 3), columns=['A', 'B', 'C']).plot.area()
```
